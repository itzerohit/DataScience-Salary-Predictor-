```python
X = df[['experience_level', 'employment_type', 'company_location', 'company_size', 'employee_residence']]
y = df['salary']
```

```python
X
```

```python
from sklearn import preprocessing
```

```python
labelencoder = preprocessing.LabelEncoder()
X["experience_level"] = labelencoder.fit_transform(X["experience_level"])
X["employment_type"] = labelencoder.fit_transform(X["employment_type"])
X["company_location"] = labelencoder.fit_transform(X["company_location"])
X["company_size"] = labelencoder.fit_transform(X["company_size"])
X["employee_residence"] = labelencoder.fit_transform(X["employee_residence"])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

def modelresults(predictions, y_test):
    print("Mean Absolute Error on model is: {}".format(mean_absolute_error(y_test, predictions)))
    print("Root Mean Squared Error on model is: {}".format(np.sqrt(mean_squared_error(y_test, predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(scaled_X_train, y_train)
```

```python
predslr = lr.predict(scaled_X_test)
```

```python
modelresults(predslr, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
```

```python
svmodel = SVR()
param_gridsvr = {"C": [0.1, 0.5], "kernel": ["linear", "rbf"], "degree": [3, 5]}
gridsvr = GridSearchCV(svmodel, param_gridsvr)
gridsvr.fit(scaled_X_train, y_train)
```

```python
print("Best parameters for the model is {}".format(gridsvr.best_params_))
```

```python
predsgridsvr = gridsvr.predict(scaled_X_test)
modelresults(predsgridsvr, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr = {"max_depth": [5,15], "max_features" : ["auto","log2"], "n_estimators" : [2,5,8,10]}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(scaled_X_train, y_train)
```

```python
print("Best parameters for the model is {}".format(gridrfr.best_params_))
```

```python
predsgridrfr = gridrfr.predict(scaled_X_test)
```

```python
modelresults(predsgridrfr, y_test)
```

```python
import pickle

# Save the trained GridSearchCV model to a file
with open("gridsvr.pkl", "wb") as f:
    pickle.dump(gridsvr, f)

print("✅ SVR GridSearchCV model saved successfully as 'gridsvr.pkl'")
```
